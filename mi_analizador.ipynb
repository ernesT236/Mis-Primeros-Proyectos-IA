```python
import pandas as pd

# Un diccionario con datos de prueba desordenados
datos = {
    "Categoria": ["Comida", "Transporte", "Entretenimiento", "Comida", "Transporte"],
    "Monto": [25.50, 12.00, 45.00, 10.20, 15.00],
}

# Creamos una tabla de datos (DataFrame)
df = pd.DataFrame(datos)

# Calculamos el total gastado por cada categoría
resumen = df.groupby("Categoria")["Monto"].sum().reset_index()

print("--- RESUMEN DE TUS GASTOS ---")
print(resumen)
```

```python
import matplotlib.pyplot as plt

# Usamos la tabla 'resumen' que creamos antes para hacer el gráfico
plt.figure(figsize=(8, 5))  # Define el tamaño de la ventana del gráfico
plt.bar(
    resumen["Categoria"], resumen["Monto"], color=["skyblue", "orange", "lightgreen"]
)

# Agregamos títulos y etiquetas para que se entienda bien
plt.title("Mis Gastos por Categoría", fontsize=14, fontweight="bold")
plt.xlabel("Categorías de Gasto", fontsize=12)
plt.ylabel("Monto Total ($)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)  # Agrega líneas de guía horizontales

# Mostramos el gráfico final en pantalla
plt.show()
```

```python
categorias = []
montos = []

print("--- INGRESA TUS GASTOS (Escribe 'salir' en categoría para terminar) ---")

while True:
    cat = (
        input("Introduce la categoría del gasto (ej. Comida, Transporte): ")
        .strip()
        .capitalize()
    )
    if cat.lower() == "salir":
        break

    try:
        monto = float(input(f"¿Cuánto gastaste en {cat}?: "))
        categorias.append(cat)
        montos.append(monto)
        print("¡Gasto guardado!")
    except ValueError:
        print("Por favor, introduce un número válido para el monto.")
    print("-" * 30)

# Guardamos lo que ingresaste en una nueva tabla limpia
mis_datos_reales = {"Categoria": categorias, "Monto": montos}
df_real = pd.DataFrame(mis_datos_reales)
resumen_real = df_real.groupby("Categoria")["Monto"].sum().reset_index()

print("\n--- DATOS PROCESADOS CON ÉXITO ---")
print(resumen_real)
```

```python
import matplotlib.pyplot as plt

# 1. Creamos el gráfico con tus nuevos datos reales
plt.figure(figsize=(8, 5))
plt.bar(
    resumen_real["Categoria"],
    resumen_real["Monto"],
    color=["#ff6666", "#66b2ff", "#99ff99"],
)

# 2. Personalizamos los textos
plt.title("Mi Reporte de Gastos Reales", fontsize=14, fontweight="bold")
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Total Gastado ($)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# 3. ¡LA MAGIA! Esta línea guarda el gráfico en tu computadora antes de mostrarlo
# Se guardará en la misma carpeta donde tienes este archivo de Jupyter
plt.savefig("reporte_gastos.png", dpi=300, bbox_inches="tight")

# 4. Lo mostramos en el cuaderno
plt.show()
```

```python
# Este código crea un archivo llamado 'gastos_banco.txt' en tu carpeta
contenido_simulado = """Comida,45.20
Transporte,12.50
Cine,25.00
Comida,15.30
Ropa,85.00
Transporte,10.00
Cine,14.50"""

with open("gastos_banco.txt", "w") as archivo:
    archivo.write(contenido_simulado.strip())

print("¡Archivo 'gastos_banco.txt' creado con éxito en tu carpeta!")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. AUTOMATIZACIÓN: Python lee el archivo de texto de forma automática
print("Leyendo archivo de gastos...")
df_auto = pd.read_csv("gastos_banco.txt", names=["Categoria", "Monto"])

# 2. PROCESAMIENTO: Agrupa y suma los gastos automáticamente
resumen_auto = df_auto.groupby("Categoria")["Monto"].sum().reset_index()
print("\n--- DATOS PROCESADOS AUTOMÁTICAMENTE ---")
print(resumen_auto)

# 3. REPORTE VISUAL: Genera y guarda el gráfico final en alta definición
plt.figure(figsize=(8, 5))
plt.bar(
    resumen_auto["Categoria"],
    resumen_auto["Monto"],
    color=["#9b59b6", "#3498db", "#e67e22", "#2ecc71"],
)

plt.title("Reporte Mensual Automatizado", fontsize=14, fontweight="bold")
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Total ($)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Guarda el nuevo reporte
plt.savefig("reporte_automatico.png", dpi=300, bbox_inches="tight")
plt.show()
print("\n¡Proceso terminado! Se ha generado 'reporte_automatico.png'.")
```
